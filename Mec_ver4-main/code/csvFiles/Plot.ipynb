```python
import pandas as pd

df = pd.read_csv("kq_full_model2.csv").values
df = df[:297964,0]
#df_slide_1000 = [df[i:(i+1000)] for i in 299]
df = df.flatten()
df_list = df.tolist()
df_list

data = pd.read_csv("FDQO_5phut_full_model.csv").values
nb_steps = data[:,5].flatten().tolist()
step = data[:,6].flatten().tolist()
# df_list_splice = [list(df_list[int(i - j): int(i)]) for j in nb_steps for i in step]
# df_list_splice
test = [df_list[int(i - j): int(i)] for j, i in zip(nb_steps, step)]
test.pop(-1) # somehow the last is empty
sum_of_subtest = [sum(sublist) / len(sublist) for sublist in test]
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import pandas as pd

plt.rcParams['figure.figsize'] = [12, 12]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower

data1 = pd.read_csv("DQL_5phut_dense.csv").values[:, 1]
data2 = pd.read_csv("DDQL_5phut_dense.csv").values[:, 1]
data3 = pd.read_csv("FDQO_5phut_dense.csv").values[:, 1]

number = [i + 1 for i in range(data1.shape[0])]
plt.tight_layout()
figure, axes = plt.subplots(nrows=3, ncols=1)


axes[0].plot(number, data1, label = 'DQL')
axes[0].plot(number, data2, label = 'DDQL')
axes[0].plot(number, data3, label = 'FDQO')
axes[0].set_title("Dense layers 450 episodes")
axes[0].legend()
axes[0].axis([0, 450, 0, 1])
axes[1].plot(number, data1, label = 'DQL')
axes[1].plot(number, data2, label = 'DDQL')
axes[1].plot(number, data3, label = 'FDQO')
axes[1].set_title("Dense layers 120 episodes")
axes[1].legend()
axes[1].axis([0, 120, 0, 1])
axes[2].plot(number, data1, label = 'DQL')
axes[2].plot(number, data2, label = 'DDQL')
axes[2].plot(number, data3, label = 'FDQO')
axes[2].set_title("Dense layers last 100 episodes")
axes[2].legend()
axes[2].axis([351, 450, 0.7, 1])
```
