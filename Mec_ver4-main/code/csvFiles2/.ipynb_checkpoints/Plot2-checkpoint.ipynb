```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import pandas as pd

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
figure, axes = plt.subplots(nrows=2, ncols=1)

data1 = pd.read_csv("DQL_5phut_1.csv").values[:, 1]
data2 = pd.read_csv("DDQL_5phut_1.csv").values[:, 1]
data3 = pd.read_csv("FDQO_5phut_1.csv").values[:, 1]
# data4 = pd.read_csv("DQL_5phut_4.csv").values[:, 1]
# data5 = pd.read_csv("DQL_5phut_5.csv").values[:, 1]

number = [i + 1 for i in range(data1.shape[0])]
plt.tight_layout()
axes[0].plot(number, data1, color="red", label = "DQL")
axes[0].plot(number, [np.average(data1[20:41])] * data1.shape[0], color="red", linestyle = '--', label = "DQL average since episode 20 - 40")
axes[0].plot(number, data2, color="green", label = "DDQL")
axes[0].plot(number, [np.average(data2[20:41])] * data1.shape[0], color="green", linestyle = '--', label = "DDQL average since episode 20 - 40")
axes[0].plot(number, data3, color="blue", label = "FDQO")
axes[0].plot(number, [np.average(data3[20:41])] * data1.shape[0], color="blue", linestyle = '--', label = "FDQO average since episode 20 - 40")
axes[0].legend()
axes[0].axis([0, 96, 0.8, 1])

df = pd.read_csv("kq_1.csv").values
df = df[:int(df.shape[0]+1),0]
#df_slide_1000 = [df[i:(i+1000)] for i in 299]
df = df.flatten()
df_list = df.tolist()
df_list

data = pd.read_csv("FDQO_5phut_1.csv").values
nb_steps = data[:,5].flatten().tolist()
step = data[:,6].flatten().tolist()
# df_list_splice = [list(df_list[int(i - j): int(i)]) for j in nb_steps for i in step]
# df_list_splice
test = [df_list[int(i - j): int(i)] for j, i in zip(nb_steps, step)]
test.pop(-1) # somehow the last is empty
sum_of_subtest = [sum(sublist) / len(sublist) for sublist in test]
axes[1].fill_between([i + 1 for i in range(len(sum_of_subtest))], 0, sum_of_subtest)
axes[1].axis([0, 96, 0, 0.4])
```
