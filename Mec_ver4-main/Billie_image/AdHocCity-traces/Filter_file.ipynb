```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

date = 8
month = 11

plt.clf()
data = pd.read_csv("data900.csv")
filt = (data['date'] == date)    
location = data.loc[filt]
filt = (location['month'] == month)   
location = location.loc[filt]
try:
    plt.scatter(location['x'], location['y'], label = '900: ' + str(len(location['x'])) +' ,'+
           str(min(location['hour'])) + ' - ' + str(max(location['hour'])))
except:
    plt.scatter(location['x'], location['y'], label = '900: ' + str(len(location['x'])))

data = pd.read_csv("data901.csv")
filt = (data['date'] == date)    
location = data.loc[filt]
filt = (location['month'] == month)   
location = location.loc[filt]
try:
    plt.scatter(location['x'], location['y'], label = '901: ' + str(len(location['x'])) +' ,'+
           str(min(location['hour'])) + ' - ' + str(max(location['hour'])))
except:
    plt.scatter(location['x'], location['y'], label = '901: ' + str(len(location['x'])))

data = pd.read_csv("data903.csv")
filt = (data['date'] == date)    
location = data.loc[filt]
filt = (location['month'] == month)   
location = location.loc[filt]
try:
    plt.scatter(location['x'], location['y'], label = '903: ' + str(len(location['x'])) +' ,'+
           str(min(location['hour'])) + ' - ' + str(max(location['hour'])))
except:
    plt.scatter(location['x'], location['y'], label = '903: ' + str(len(location['x'])))

plt.legend()
plt.title(str(date) + ' / ' + str(month))
plt.savefig('./image/' +str(date) + ' - ' + str(month) +'a.png')
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data900.csv")
a1 = np.unique(df[['date', 'month']].values, axis = 0)
a1 = a1[a1[:, 0].argsort()]
a1 = a1[a1[:, 1].argsort(kind='mergesort')]
# print(a1)
# print()

df = pd.read_csv("data901.csv")
a2 = np.unique(df[['date', 'month']].values, axis = 0)
a2 = a2[a2[:, 0].argsort()]
a2 = a2[a2[:, 1].argsort(kind='mergesort')]
# print(a2)
# print()

df = pd.read_csv("data903.csv")
a3 = np.unique(df[['date', 'month']].values, axis = 0)
a3 = a3[a3[:, 0].argsort()]
a3 = a3[a3[:, 1].argsort(kind='mergesort')]
# print(a3)
# print()
np.intersect1d( np.intersect1d(a1, a3), a2 )
```
